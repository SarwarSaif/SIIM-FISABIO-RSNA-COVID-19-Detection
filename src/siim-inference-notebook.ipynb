```python
import numpy as np, pandas as pd
from glob import glob
import shutil, os
import matplotlib.pyplot as plt
from sklearn.model_selection import GroupKFold
from tqdm.notebook import tqdm
import seaborn as sns
```

```python
CNV_IMG_PATH = "../input/siim-covid19-resized-to-256px-jpg/"
DF_PATH = "../input/siimcovid19traindf/train_df.csv"
KDF_PATH = "../input/siimcovid19traindf/kfold_df.csv"
```

```python
train_df = pd.read_csv(f'{DF_PATH}')
train_df.head()
```

```python
train_df['path'][0]
```

```python
from sklearn.model_selection import GroupKFold
kfold_df = train_df.copy()

gkf  = GroupKFold(n_splits = 5)
kfold_df['fold'] = -1
for fold, (train_idx, val_idx) in enumerate(gkf.split(kfold_df, groups = kfold_df.StudyInstanceUID.tolist())):
    kfold_df.loc[val_idx, 'fold'] = fold

kfold_df.head()
```

```python
kfold_df['ImageID'] = kfold_df['id'].apply(lambda x: x.replace('_image', ''))
```

```python
kfold_df.head()
```

```python
# Create Training Paths
train_directory = f"{CNV_IMG_PATH}train/"
kfold_df['image_path'] = f'{CNV_IMG_PATH}train/' + kfold_df.ImageID + '.jpg'
kfold_df.head()
```

```python
kfold_df.to_csv("kfold_df.csv", index=True)
```
